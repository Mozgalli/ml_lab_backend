## 1. VektÃ¶rler

VektÃ¶rler, makine Ã¶ÄŸreniminde veri noktalarÄ±nÄ± temsil eder.

```python
# VektÃ¶r oluÅŸturma
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print(f"v1 = {v1}")
print(f"v2 = {v2}")

# VektÃ¶r toplama
v_sum = v1 + v2
print(f"\nv1 + v2 = {v_sum}")

# VektÃ¶r normu (uzunluÄŸu)
norm_v1 = np.linalg.norm(v1)
print(f"\n||v1|| = {norm_v1:.2f}")

# Ä°Ã§ Ã§arpÄ±m (dot product)
dot_product = np.dot(v1, v2)
print(f"\nv1 Â· v2 = {dot_product}")
```

## 2. Matrisler

Matrisler, birden fazla vektÃ¶rÃ¼ veya dÃ¶nÃ¼ÅŸÃ¼mleri temsil eder.

```python
# Matris oluÅŸturma
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matris A:")
print(A)
print("\nMatris B:")
print(B)

# Matris Ã§arpÄ±mÄ±
C = np.dot(A, B)
print("\nA Ã— B =")
print(C)

# Matris transpozu
A_T = A.T
print("\nAáµ€ (transpose) =")
print(A_T)
```

## 3. Determinant ve Ters Matris

Determinant, matrisin 'hacmini' Ã¶lÃ§er. Ters matris, lineer denklem sistemlerini Ã§Ã¶zmek iÃ§in kullanÄ±lÄ±r.

```python
# Determinant
det_A = np.linalg.det(A)
print(f"det(A) = {det_A}")

# Ters matris
A_inv = np.linalg.inv(A)
print("\nAâ»Â¹ (inverse) =")
print(A_inv)

# DoÄŸrulama: A Ã— Aâ»Â¹ = I (birim matris)
identity = np.dot(A, A_inv)
print("\nA Ã— Aâ»Â¹ = I (birim matris):")
print(np.round(identity, 10))  # Yuvarlama (floating point hatalarÄ±nÄ± gizlemek iÃ§in)
```

## 4. Eigenvalues ve Eigenvectors

Ã–z deÄŸerler ve Ã¶z vektÃ¶rler, matrisin 'Ã¶zel yÃ¶nlerini' gÃ¶sterir.

**TanÄ±m**: Av = Î»v (Î»: eigenvalue, v: eigenvector)

```python
# Eigenvalues ve eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues (Î»):")
print(eigenvalues)
print("\nEigenvectors (v):")
print(eigenvectors)

# DoÄŸrulama: Av = Î»v
v1 = eigenvectors[:, 0]
lambda1 = eigenvalues[0]

Av1 = np.dot(A, v1)
lambda_v1 = lambda1 * v1

print("\nDoÄŸrulama (ilk eigenvector iÃ§in):")
print(f"Av = {Av1}")
print(f"Î»v = {lambda_v1}")
print(f"EÅŸit mi? {np.allclose(Av1, lambda_v1)}")
```

## 5. GÃ¶rselleÅŸtirme: VektÃ¶r Transformasyonu

```python
# Orijinal vektÃ¶r
v = np.array([1, 1])

# Transformasyon matrisi (rotation + scaling)
T = np.array([[2, 0], [0, 1]])

# Transform edilmiÅŸ vektÃ¶r
v_transformed = np.dot(T, v)

# GÃ¶rselleÅŸtirme
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.006, label='Original')
plt.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.006, label='Transformed')

plt.xlim(-0.5, 3)
plt.ylim(-0.5, 2)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.legend()
plt.title('VektÃ¶r Transformasyonu')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print(f"Orijinal vektÃ¶r: {v}")
print(f"Transform edilmiÅŸ: {v_transformed}")
```
